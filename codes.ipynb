```python
def count_characters(text):
    vowels = "aeiouAEIOU"

    v_count = 0
    c_count = 0
    d_count = 0
    s_count = 0

    for char in text:
        if char.isalpha():
            if char in vowels:
                v_count += 1
            else:
                c_count += 1
        elif char.isdigit():
            d_count += 1
        else:
            s_count += 1

    print("Vowels:", v_count)
    print("Consonants:", c_count)
    print("Digits:", d_count)
    print("Special Characters:", s_count)



text = input("Enter a string: ")
count_characters(text)
```

```python
def find_second_elements(lst):
    if len(lst) < 2:
        print("List must have at least two elements.")
        return


    largest = second_largest = float('-inf')
    smallest = second_smallest = float('inf')

    for num in lst:

        if num > largest:
            second_largest = largest
            largest = num
        elif num > second_largest and num != largest:
            second_largest = num


        if num < smallest:
            second_smallest = smallest
            smallest = num
        elif num < second_smallest and num != smallest:
            second_smallest = num


    if second_largest == float('-inf'):
        print("No second largest element (all elements may be equal).")
    else:
        print("Second Largest:", second_largest)

    if second_smallest == float('inf'):
        print("No second smallest element (all elements may be equal).")
    else:
        print("Second Smallest:", second_smallest)



lst = [10, 20, 4, 45, 99, 99, 20]
print("List:", lst)
find_second_elements(lst)
```

```python
def remove_duplicates(lst):
    result = []

    for item in lst:
        if item not in result:
            result.append(item)

    return result



lst = [10, 20, 10, 30, 20, 40, 30]
print("Original List:", lst)

new_list = remove_duplicates(lst)
print("After Removing Duplicates:", new_list)
```

```python
students = {
    "Aman": 85,
    "Riya": 92,
    "Karan": 78,
    "Sneha": 95,
    "Neha": 88
}

print("Student Data:", students)


topper = None
highest_marks = -1

for name, marks in students.items():
    if marks > highest_marks:
        highest_marks = marks
        topper = name

print("Topper:", topper, "-", highest_marks)


total = 0
count = 0

for marks in students.values():
    total += marks
    count += 1

average = total / count
print("Average Marks:", average)


sorted_students = sorted(students.items(), key=lambda x: x[1])

print("Students Sorted by Marks:")
for name, marks in sorted_students:
    print(name, ":", marks)
```

```python
def word_frequency(text):
    freq = {}

    words = text.lower().split()

    for word in words:

        word = word.strip(".,!?;:'\"()[]{}")

        if word in freq:
            freq[word] += 1
        else:
            freq[word] = 1

    return freq



text = input("Enter a text: ")

result = word_frequency(text)

print("\nWord Frequencies:")
for word, count in result.items():
    print(word, ":", count)
```

```python
def detect_and_convert(user_input):
    # Try integer
    try:
        value = int(user_input)
        return value, type(value)
    except ValueError:
        pass

    # Try float
    try:
        value = float(user_input)
        return value, type(value)
    except ValueError:
        pass

    # Try boolean
    if user_input.lower() in ['true', 'false']:
        value = user_input.lower() == 'true'
        return value, type(value)

    # Default: string
    return user_input, type(user_input)



user_input = input("Enter any value: ")

value, datatype = detect_and_convert(user_input)

print("Converted Value:", value)
print("Detected Data Type:", datatype)
```

```python
def is_palindrome_no_slice(s):
    s = s.lower()
    left = 0
    right = len(s) - 1

    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1

    return True

# Input from user
text = input("Enter a string: ")

if is_palindrome_no_slice(text):
    print("Palindrome")
else:
    print("Not a palindrome")
```
